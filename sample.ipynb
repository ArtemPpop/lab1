```python
array = [
    [3, 5, 7, 2],
    [1, 4, 6, 3],
    [2, 8, 5, 1]
]

rows = len(array)
cols = len(array[0])

min_sum = float('inf')
min_index = 0

for j in range(cols - 1):
    s = 0
    for i in range(rows):
        s += array[i][j] + array[i][j + 1]
    if s < min_sum:
        min_sum = s
        min_index = j

print("Минимальная сумма между столбцами", min_index, "и", min_index + 1, "=", min_sum)
```

```python
num = [1, 2, 3, 4, 5]
k = 1
C = 99

num.append(None)
for i in range(len(num) - 1, k, -1):
    num[i] = num[i - 1]
num[k] = C

print(num)
```

```python
def superset(A, B):
    if A == B:
        print("Множества равны")
    elif A.issuperset(B):
        print(f" {A} является множеством")
    elif B.issuperset(A):
        print(f" {B} является множеством")
    else:
        print("множество не обнаружено")

superset({1, 2, 3}, {1, 2})
```

```python
cats = [
    ("Мурзик", 3, "Анна", "Петрова"),
    ("Барсик", 5, "Игорь", "Сидоров")
]

cat_dict = {cat[0]: {"Возраст": cat[1], "Имя владельца": cat[2], "Фамилия владельца": cat[3]} for cat in cats}

for name, info in cat_dict.items():
    print(name, "-", info)
```
